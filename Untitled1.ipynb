```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile('Test_dataset.xlsx')
df = pd.read_excel(xls, 'Put-Call_TS')
```

```python
df.head()
```

```python
df.columns = ['Stock Index','10-Aug','11-Aug','12-Aug','13-Aug','14-Aug','15-Aug']
df.head()
```

```python
df = df.iloc[1:]
df.head()
```

```python
df.info()
```

```python
a = ['10-Aug','11-Aug','12-Aug','13-Aug','14-Aug','15-Aug']

for i in a:
    df[i].fillna(df[i].mean(), inplace = True) 
    
df.info()
```

```python
a = df.iloc[1:2,1:7]
print(a)
r = [a['10-Aug'], a['11-Aug'], a['12-Aug'], a['13-Aug'], a['14-Aug'], a['15-Aug']]
print(r[5])
```

```python
def append(row):
    arr = []
    arr.append(row['10-Aug'])
    arr.append(row['11-Aug'])
    arr.append(row['12-Aug'])
    arr.append(row['13-Aug'])
    arr.append(row['14-Aug'])
    arr.append(row['15-Aug'])
    return arr

for index,row in df.head(2).iterrows():
    a = append(row)
    print(a)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

result = []

for index,row in df.iterrows():
    data = append(row)
    model = SimpleExpSmoothing(data)
    model_fit = model.fit()
    # make prediction
    yhat = model_fit.predict(len(data), len(data))
    result.append(yhat[-1])
    
result
```

```python
with open("part3.txt", "a") as f:
    for i in result:
        f.write(i.astype(str)+"\n")
```
